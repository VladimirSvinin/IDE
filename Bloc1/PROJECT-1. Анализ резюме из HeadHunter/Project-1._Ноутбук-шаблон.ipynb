* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def work_experience_months (work_experience):
    if work_experience is np.nan: pass
    elif work_experience == "Не указано": work_experience = np.nan
    else:
        work_experience_split = work_experience.split()[:6]
        condition1 = "месяц" in work_experience_split or "месяца" in work_experience_split or "месяцев" in work_experience_split
        condition2 = "лет" in work_experience_split or "год" in work_experience_split or "года" in work_experience_split
        if condition1 and not condition2:
            work_experience_split = work_experience.split()[:4]
            work_experience = int(work_experience_split[2])
        elif condition2 and not condition1:
            work_experience_split = work_experience.split()[:4]
            work_experience = int(work_experience_split[2]) * 12
        elif condition1 and condition2:
            work_experience = int(work_experience_split[2]) * 12 + int(work_experience_split[4])
    return work_experience
job_data['Опыт работы (мес.)'] = job_data['Опыт работы'].apply(work_experience_months)
#ваш код здесь
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
# Часть 1 - определяем город
def city_extract (city_string):
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    city = city_string.split(',')[0]
    city = city.strip()
    if city == "Москва": city = "Москва"
    elif city == "Санкт-Петербург": city = "Санкт-Петербург"
    elif city in million_cities: city = "город-миллионник"
    else: city = "другие"
    return city
#city_extract("Екатеринбург , готов к переезду (Сочи, Москва, Санкт-Петербург) , готов к командировкам")
job_data['Город'] = job_data['Город, переезд, командировки'].apply(city_extract)
job_data['Город'].value_counts() # - функция работает правильно

#Часть 2 - определяем готовность к переезду
def ready_to_move (ready_string):
    ready_string_split = ready_string.split(',')
    condition1 = "м. " in ready_string_split[1] #проверяем, есть указана ли в строке информация о метро
    if condition1: # если указана, то удаляем её
        ready_string_split.remove(ready_string_split[1])
    if "готов" in ready_string_split[1] and "не" in ready_string_split[1]: return False #теперь проверяем, чтобы в искомой строке были одновременно были "не" и "готов"
    #такую строку будем считать за нежелание переезда
    else: return True
#print(ready_to_move('Воронеж , не готова к переезду (Сочи, Москва, Санкт-Петербург) , готов к командировкам'))
#print(ready_to_move('Москва , м. Беломорская , готова к переезду, не готов к командировкам'))
#print(ready_to_move('Воронеж , не готова к переезду (Сочи, Москва, Санкт-Петербург) , готов к командировкам'))
#cудя по всему всё работает правильно
job_data['Готовность к переезду'] = job_data['Город, переезд, командировки'].apply(ready_to_move)
#job_data['Готовность к переезду'].value_counts()

#Часть 3 - определяем готовность к коммандировкам
def ready_for_trips(trips_string):
    ready_for_trips_split = trips_string.split(',')
    workable_string = ready_for_trips_split[-1] #работаем с последней строкой в списке
    if "готов" in workable_string and "не" in workable_string: return False
    else: return True
#print(ready_for_trips('Воронеж , готов к переезду (Сочи, Москва, Санкт-Петербург) , готов к командировкам'))
#print(ready_for_trips('Москва , м. Беломорская , не готов к переезду, не готов к командировкам'))
#print(ready_for_trips('Воронеж , готов к переезду (Сочи, Москва, Санкт-Петербург) , не готова к командировкам'))
#функция работает правильно
job_data['Готовность к командировкам'] = job_data['Город, переезд, командировки'].apply(ready_for_trips)
job_data['Готовность к командировкам'].value_counts()
job_data.drop(axis = 1, labels = 'Город, переезд, командировки')
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
#создадим по функции для каждого вида занятости и графика
def full_day(str):
    if "полная занятость" in str: return True
    else: return False
def part_time(str):
    if "частичная занятость" in str: return True
    else: return False
def project(str):
    if "проектная работа" in str: return True
    else: return False
def volunteer (str):
    if "волонтерство" in str: return True
    else: return False
def internship (str):
    if "стажировка" in str: return True
    else: return False
def full_day (str):
    if "полный день" in str: return True
    else: return False
def shifts (str):
    if "сменный график" in str: return True
    else: return False
def flexible (str):
    if "гибкий график" in str: return True
    else: return False
def remote_work (str):
    if "удаленная работа" in str: return True
    else: return False
def vahta (str):
    if "вахтовый метод" in str: return True
    else: return False

#print(vahta('вахтовый метод, гибкий график, удаленная работа, полная занятость'))
#print(flexible('вахтовый метод, гибкий график, удаленная работа, полная занятость'))
#print(internship('вахтовый метод, гибкий график, удаленная работа, полная занятость'))
#print(volunteer('вахтовый метод, гибкий график, удаленная работа, полная занятость'))
#функция работает правильно

job_data['полная занятость'] = job_data['Занятость'].apply(full_day)
job_data['частичная занятость'] = job_data['Занятость'].apply(part_time)
job_data['проектная работа'] = job_data['Занятость'].apply(project)
job_data['волонтерство'] = job_data['Занятость'].apply(volunteer)
job_data['стажировка'] = job_data['Занятость'].apply(internship)
job_data['полный день'] = job_data['График'].apply(full_day)
job_data['сменный график'] = job_data['График'].apply(shifts)
job_data['гибкий график'] = job_data['График'].apply(flexible)
job_data['удаленная работа'] = job_data['График'].apply(remote_work)
job_data['вахтовый метод'] = job_data['График'].apply(vahta)
job_data[(job_data['вахтовый метод'] == True) & (job_data['гибкий график'] == True)].shape[0]
job_data = job_data.drop(axis = 1, labels = ['График','Занятость'])
job_data
```

Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь
exchange_rates = pd.read_csv('data/ExchangeRates.csv') # подргружаем csv с курсами валют
exchange_rates['date'] = pd.to_datetime(exchange_rates['date'], dayfirst = True) #переводим даты в формат datetime
exchange_rates = exchange_rates.drop(axis = 1, labels = ['per', 'time', 'vol']) #удаляем все ненужные столбцы из таблицы

#работаем с основным data frame
job_data['Обновление резюме'] = pd.to_datetime(job_data['Обновление резюме'], dayfirst = True).dt.date
job_data['Обновление резюме'] = pd.to_datetime(job_data['Обновление резюме'], dayfirst = True) #Переводим данные в формат datetime и оставляем только саму дату
#создадим 2 функции для того, чтобы создать 2 новых признака в таблице job_data - валюта и размер ЗП
def salary_extract (salary):
    salary = int(salary.split()[0])
    return salary
def currency_extract(salary):
    currency = salary.split()[1]
    return currency
job_data['Валюта'] = job_data['ЗП'].apply(currency_extract)
job_data['З/П'] = job_data['ЗП'].apply(salary_extract)
# проверим, как прошло извлечение
#job_data['Валюта'].value_counts() # всё высвечивается корректно, никаких ошибок нет
#job_data['З/П'].value_counts()

#Переводим валюты из job_data['Валюта'] в формат ISO
def currency_iso_convert(currency):
    if currency == 'руб.': currency = 'RUB'
    elif currency == 'KZT': pass
    elif currency == 'USD': pass
    elif currency == 'бел.руб.': currency = 'BYN'
    elif currency == 'EUR': pass
    elif currency == 'грн.': currency = 'UAH'
    elif currency == 'сум': currency = 'UZS'
    elif currency == 'AZN': pass
    return currency
job_data['Валюта'] = job_data['Валюта'].apply(currency_iso_convert) #Валюты правильно переведены в формат ISO
#job_data['Валюта'].value_counts()

#Присоединяем таблицы друг к другу
merged_table = job_data.merge(
    exchange_rates,
    left_on = ["Обновление резюме", "Валюта"],
    right_on = ["date", "currency"],
    how = "left"
)

#Заполняем пропуски в новой таблице:
empty_values_dict = {"close" : 1, "proportion" : 1}
merged_table = merged_table.fillna(empty_values_dict)

# пересчитываем все ЗП в рублях, создадим для этого новый признак
merged_table['Итоговая ЗП (руб.)'] = merged_table['З/П'] / merged_table['proportion'] * merged_table['close']
merged_table['Итоговая ЗП (руб.)'].median()
# Удаляем ненужные столбцы
merged_table = merged_table.drop(axis = 1, labels= ['ЗП', 'Город, переезд, командировки', 'proportion', 'currency', 'date','close', 'Валюта', 'З/П'])
```
